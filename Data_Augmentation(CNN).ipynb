## Creating Image Data Generator

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# creating Image Data Generator function
datagen = ImageDataGenerator(
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True)
```

```python
data = img_to_array(data)
data
```

```python
data.shape
```

```python
# Data reshape
input_data = data.reshape(1,200,200,3)
```

```python
i = 0
for output in datagen.flow(input_data,batch_size=1,save_to_dir='/content/drive/MyDrive/Colab Notebooks/Deep Learning/Learn from outside'):

  i = i+1

  if i == 10:
    break
```
